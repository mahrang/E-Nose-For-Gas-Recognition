# DATA SET CLEANING & VISUALIZATION 



```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from matplotlib import pyplot as plt
%matplotlib inline
```

# Reading Raw Dataset

```python
df_1 = pd.read_excel('batch1.xlsx', header=None)
df_2 = pd.read_excel('batch2.xlsx', header=None)
df_3 = pd.read_excel('batch3.xlsx', header=None)
df_4 = pd.read_excel('batch4.xlsx', header=None)
df_5 = pd.read_excel('batch5.xlsx', header=None)
df_6 = pd.read_excel('batch6.xlsx', header=None)
df_7 = pd.read_excel('batch7.xlsx', header=None)
df_8 = pd.read_excel('batch8.xlsx', header=None)
df_9 = pd.read_excel('batch9.xlsx', header=None)
df_10 = pd.read_excel('batch10.xlsx', header=None)
```

# Cleaning The Data
we can drop every other column as it contains no information about the sensors.

```python
batch1_clean = df_1[df_1.columns[0::2]]
batch2_clean = df_2[df_2.columns[0::2]]
batch3_clean = df_3[df_3.columns[0::2]]
batch4_clean = df_4[df_4.columns[0::2]]
batch5_clean = df_5[df_5.columns[0::2]]
batch6_clean = df_6[df_6.columns[0::2]]
batch7_clean = df_7[df_7.columns[0::2]]
batch8_clean = df_8[df_8.columns[0::2]]
batch9_clean = df_9[df_9.columns[0::2]]
batch10_clean = df_10[df_10.columns[0::2]]
```

```python
Data_vis = [batch1_clean, batch2_clean, batch3_clean, batch4_clean, batch5_clean,
            batch6_clean, batch7_clean, batch8_clean, batch9_clean, batch10_clean]
```

```python
fig = plt.figure(figsize = (50,75))

for i in range(len(Data_vis)):
    
    # adding new batch to V every iteration.
    V = pd.concat(Data_vis[0:i+1])
    
    # resetting the index, as every new batch added has its own index
    V = V.reset_index()
    V = V.drop(['level_0'], axis=1)
    V = V.drop(['index'], axis=1)
    V.index = V.index + 1
    
    # taking all 128 features in X_vis
    X_vis = V[V.columns[1:]]
    # Standardizing the all the features
    X_vis = StandardScaler().fit_transform(X_vis)
    # taking output in y_vis
    y_vis = V[V.columns[0:1]]

    #Using PCA to reduce the features to have better visualization of the data set
    pca = PCA(n_components=2)
    principalComponents = pca.fit_transform(X_vis)
    # taking both Principal components in a new data frame, principalDF
    principalDf = pd.DataFrame(data = principalComponents, columns = ['Principal component 1', 'Principal component 2'])
    
    # adding the output column to the principal components data frame    
    finalDf = pd.concat([principalDf, y_vis], axis = 1)
    finalDf.columns = ['Principal component 1', 'Principal component 2', 'Y']
    

    Visualization = fig.add_subplot(5,2,(i+1)) 
    Visualization.set_xlabel('Principal Component 1', fontsize = 30)
    Visualization.set_ylabel('Principal Component 2', fontsize = 30)

    # Mapping all 6 classes to different colors
    targets = [1,2,3,4,5,6]
    colors = ['b', 'g', 'r', 'c', 'm', 'y']
    # ploting all the points with colors mapped above
    for target, color in zip(targets,colors):
        indicesToKeep = finalDf['Y'] == target
        Visualization.scatter(finalDf.loc[indicesToKeep, 'Principal component 1'],
                              finalDf.loc[indicesToKeep, 'Principal component 2'],
                              c = color,
                              s = 140)
        if i == 0:
            Visualization.set_title('Plot for batch %i'%(i+1), fontsize = 40)
        else:
            Visualization.set_title('Plot for first %i batches'%(i+1), fontsize = 40)
    Visualization.legend(targets, prop={'size': 21})
    plt.axis([-20, 40, -30, 30])
    Visualization.grid()
    
```

It is clear from the plot that as we add more batches we see the classes mixing with each other(effect of sensor drift).
